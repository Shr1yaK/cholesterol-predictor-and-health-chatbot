```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    tokens = word_tokenize(text.lower())
    tokens = [lemmatizer.lemmatize(token) for token in tokens if token.isalpha()]
    tokens = [token for token in tokens if token not in stop_words]
    return ' '.join(tokens)

preprocessed_dataset = [(preprocess_text(question), response) for question, response in dataset] 
```

```python
X_train = [question for question, _ in preprocessed_dataset]
y_train = [response for _, response in preprocessed_dataset]

vectorizer = TfidfVectorizer()
X_train_vectors = vectorizer.fit_transform(X_train)

 
model=MultinomialNB()
model = SVC(probability=True)  
model.fit(X_train_vectors, y_train)

X_train_vectors = vectorizer.transform(X_train)
train_accuracy = model.score(X_train_vectors, y_train)
print("Model Accuracy: {:.2f}".format(train_accuracy))
```

```python
user_name = input("Please enter your name: ")
print("EudemoniaBot: Hi", user_name + "! How can I assist you today?")

while True:
    user_input = input(user_name + ": ")
    preprocessed_input = preprocess_text(user_input)
    input_vector = vectorizer.transform([preprocessed_input])
    predicted_response = model.predict(input_vector)[0]
    similarities = cosine_similarity(input_vector, X_train_vectors)
    max_similarity = similarities.max()
    confidence_level = max_similarity.item()
    print(confidence_level)
    if confidence_level < 0.6:
        print("EudemoniaBot: As an AI, I don't have an answer for that. Please send an email to gowri.kishorek@gmail.com.")
    else:
        print("EudemoniaBot:", predicted_response)
    additional_question = input("EudemoniaBot: Is there anything else you want to know, " + user_name + "? (yes/no): ")
    
    if additional_question.lower() == "no":
        print("EudemoniaBot: Okay, " + user_name + ", have a great day!")
        feedback = input("EudemoniaBot: Please provide your feedback, " + user_name + ": ")
        sentiment = TextBlob(feedback).sentiment
        polarity = sentiment.polarity
        
        if polarity > 0:
            print("EudemoniaBot: I'm glad you found my response helpful, " + user_name + "!") 
            break
        elif polarity < 0:
            print("EudemoniaBot: I apologize if my response was not satisfactory, " + user_name + ". How can I improve?")
            feedbackback_input = input("User: ")
            print("EudemoniaBot: Thank you for your feedback!")
            break
        else:
            print("EudemoniaBot: Thank you for your feedback, " + user_name + "!")
            break
    elif additional_question.lower() == "yes":
        continue
```
